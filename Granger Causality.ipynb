# Granger Causality Testing

### Closing Price vs Number of Articles

```python
# These are the two series for input into our granger causality test
x = np.asarray(btc_2017[['close', 'num_articles']])

grangercausalitytests(x, 10, addconst=True, verbose=True)

# We find predictions of the closing price are indeed improved upon considering past values of the numArticles variable
# as opposed to predictions based on just past values of the closing price itself.
```

```python
# Checking other direction

x = np.asarray(btc_2017[['num_articles', 'close']])

grangercausalitytests(x, 10, addconst=True, verbose=True)

# We find that there is even stronger evidence that the number of articles depends upon the
# closing price from previous periods. 
```

As expected we have found bi-directional granger causality between the closing price of bitcoin and the number of articles tagged as pertaining to bitcoin. In our previous tests we had not found this to be the case. I believe that we were passing in differenced and transformed data to a method which performs these operations on the raw data. Perhaps double differencing and double transformations were the cause of our previous test results.

### Closing Price vs Normalised Tone

```python
# These are the two series for input into our granger causality test
x = np.asarray(btc_2017[['close', 'tone']])

grangercausalitytests(x, 10, addconst=True, verbose=True)

# Here we find that the tone series gives us no extra predictive power when trying to predict values for closing price
```

```python
# Checking other direction

x = np.asarray(btc_2017[['tone', 'close']])

grangercausalitytests(x, 10, addconst=True, verbose=True)

# We find again no granger causality in the other direction 
```
